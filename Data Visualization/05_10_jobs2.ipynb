```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../data/occupations.csv")
df
```

```python
occu_count = df.groupby('occupation').count().iloc[:, 0]
occu_count
```

```python
occu_F = df[df['gender']=='F'].groupby('occupation').count().iloc[:, 0]
```

```python
combined = pd.concat([occu_count, occu_F], axis=1)
combined = combined.fillna(0)
combined
```

```python
combined = combined.iloc[:,1]/combined.iloc[:,0]
combined
```

```python
combined.sort_values(ascending=False)
```

```python
# 모범답안
df = pd.read_csv('../data/occupations.csv')

# 여기에 코드를 작성하세요
occupation_group = df.groupby('occupation')

df.loc[df['gender'] == 'M', 'gender'] = 0
df.loc[df['gender'] == 'F', 'gender'] = 1

occupation_group.mean()['gender'].sort_values(ascending=False)
```
