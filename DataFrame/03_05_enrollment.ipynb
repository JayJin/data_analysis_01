```python
import pandas as pd
```

```python
df = pd.read_csv("../data/enrolment_1.csv")
df.head()
```

```python
df['status'] = 'allowed'
df.head()
```

```python
# conditions
c_1 = df['course name'] == 'information technology'
c_2 = df['year'] == 1
c_3 = df['course name'] == 'commerce'
c_4 = df['year'] == 4
c_5 = df['course name'].value_counts() < 5
```

```python
df.loc[c_1 & c_2, 'status'] = 'not allowed'
df.loc[c_3 & c_4, 'status'] = 'not allowed'
```

```python
df.loc[df['course name'].isin(c_5[c_5].index), 'status'] = 'not allowed'
```

```python
df['room assignment'] = 'not assigned'
df.head()
```

```python
c_6 = (df['course name'].value_counts() >= 80)
c_7 = (df['course name'].value_counts() >= 40)&(df['course name'].value_counts() < 80)
c_8 = (df['course name'].value_counts() >= 15)&(df['course name'].value_counts() < 40)
c_9 = (df['course name'].value_counts() >= 5)&(df['course name'].value_counts() < 15)
c_10 = df['status'] == 'not allowed'
```

```python
df.loc[df['course name'].isin(c_6[c_6].index), 'room assignment'] = 'Auditorium'
df.loc[df['course name'].isin(c_7[c_7].index), 'room assignment'] = 'Large room'
df.loc[df['course name'].isin(c_8[c_8].index), 'room assignment'] = 'Medium room'
df.loc[df['course name'].isin(c_9[c_9].index), 'room assignment'] = 'Small room'
df.loc[c_10, 'room assignment'] = 'not assigned'
```

```python
df.head()
```

```python
room_aud = list(df.loc[df['room assignment']=='Auditorium', 'course name'].unique())
room_lar = list(df.loc[df['room assignment']=='Large room', 'course name'].unique())
room_med = list(df.loc[df['room assignment']=='Medium room', 'course name'].unique())
room_sma = list(df.loc[df['room assignment']=='Small room', 'course name'].unique())
room_aud.sort()
room_lar.sort()
room_med.sort()
room_sma.sort()
```

```python
cnt = 1
for i in room_aud:
    df.loc[(df['course name'] == i)&(df['room assignment']!='not assigned'), 'room assignment'] = f'Auditorium-{cnt}'
    cnt += 1

cnt = 1
for i in room_lar:
    df.loc[(df['course name'] == i)&(df['room assignment']!='not assigned'), 'room assignment'] = f'Large-{cnt}'
    cnt += 1

cnt = 1
for i in room_med:
    df.loc[(df['course name'] == i)&(df['room assignment']!='not assigned'), 'room assignment'] = f'Medium-{cnt}'
    cnt += 1

cnt = 1
for i in room_sma:
    df.loc[(df['course name'] == i)&(df['room assignment']!='not assigned'), 'room assignment'] = f'Small-{cnt}'
    cnt += 1
df
```

```python
df.rename(columns={'room assignment':'room number'}, inplace=True)
df
```
